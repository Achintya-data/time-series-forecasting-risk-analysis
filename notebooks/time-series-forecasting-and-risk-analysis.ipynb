**Differencing + ADF Retest**

```r
# First Differencing for Baltic Dry Index (BDI) ---
bdi_diff <- diff(bdi_ts, differences = 1)

# Plot differenced BDI series
autoplot(bdi_diff) +
  labs(
    title = "Differenced Baltic Dry Index (1st Difference)",
    x = "Year",
    y = "Differenced BDI Price"
  ) +
  theme_minimal()

# ADF Test on Differenced BDI
adf.test(bdi_diff)
```

For Baltic Dry Index (BDI)


After applying first-order differencing to the Baltic Dry Index (BDI) data, the resulting time series appears visually more stable around a constant mean with reduced variance. The differenced plot shows no obvious trends or long-term fluctuations, indicating successful stationarity transformation. This step was necessary to prepare the BDI series for ARIMA model building.

```r
# First Order Differencing
ice_cream_diff1 <- diff(ice_cream_ts, differences = 1)

# Seasonal Differencing (Lag-12)
ice_cream_diff_seasonal <- diff(ice_cream_diff1, lag = 12)

# Plot the final differenced series
autoplot(ice_cream_diff_seasonal) +
  labs(
    title = "First and Seasonal Differenced Ice Cream Production Time Series",
    x = "Year",
    y = "Differenced Production Index"
  ) +
  theme_minimal()

# ADF Test on the Final Differenced Series
adf.test(ice_cream_diff_seasonal)
```

For Ice Cream Industrial Production Index


After applying both first-order and seasonal differencing to the Ice Cream Production Index data, the resulting time series appears visually stabilized with reduced trend and seasonality. The differenced plot shows a consistent mean and variance over time, confirming successful stationarity transformation. This preparation was essential for fitting an appropriate SARIMA model to the data.

### How to Interpret ACF and PACF:

| Pattern Seen | Suggests |
|:-------------|:---------|
| ACF cuts off after lag k | MA(k) model |
| PACF cuts off after lag p | AR(p) model |
| ACF and PACF both tail off slowly | ARMA model needed |
| Significant spikes at multiples of seasonal lag (12, 24, etc.) | Need Seasonal ARIMA (SARIMA) |

**Model Fitting and Diagnostics**

| Diagnostic | What it Tells Us|
|-----------------|-----------------------------------|
|Residual plot    | Residuals should look like white noise (no patterns)|
|ACF of residuals | No significant autocorrelations should remain|
|Ljung-Box test   | Tests if residuals are random (p-value > 0.05 desired)|

Interpretation:

* Residuals look random + Ljung-Box p-value > 0.05	 :  Model is a good fit

* Residuals show pattern or Ljungbox p-value < 0.05  :  Model might need improvements

Why the Point Forecast is Constant for BDI?:

* We are using an ARIMA(0,1,5) model.

* ARIMA(0,1,5) has only differencing I(1) + Moving Average terms MA(5).

* ARIMA models without AR or seasonal components behave like random walks after differencing.

* When forecast into the future:

  1. The best estimate (mean forecast) flattens into the last known value.

  2. Uncertainty grows (hence confidence intervals widen), but the mean stays constant.

* It is normal behavior for pure MA models and random walks.

The Baltic Dry Index (Non-Seasonal) forecast using ARIMA(0,1,5) displays a flat point forecast line.

This behavior is expected because ARIMA(0,1,5) represents a random walk model, where the best estimate for future values remains the last observed value, and uncertainty grows over time.

The widening confidence intervals reflect increasing forecast uncertainty.
